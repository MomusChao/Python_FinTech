```python
import twder
```

```python
# print currency code
twder.currencies()
```

```python
# current status
# (time,Cash/Bank Buy, Cash/Bank Sell, Spot/Bank Buy, Spot/Bank Sell)
twder.now_all()
```

```python
#print one currency info
twder.now('USD')
```

```python
# print USD Cash/Bank Sell
USD = float(twder.now('USD')[2])
USD
```

```python
# yesterday status
twder.past_day('USD')
```

```python
# last 6 months status
twder.past_six_month('USD')
```

```python
# specify month
twder.specify_month('USD', 2018, 12)
```
